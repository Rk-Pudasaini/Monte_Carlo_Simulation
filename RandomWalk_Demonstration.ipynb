# What is Random Walk?
A random walk is a mathematical concept that describes a path or trajectory formed by a sequence of random steps. Each step is taken in a random direction, often with equal probability of moving in different directions. Random walks are widely used to model various phenomena in fields such as physics, finance, biology, and computer science.

In its simplest form, a random walk can occur in one dimension (1D), where the walker moves along a straight line either to the left or right with equal probability at each step. This can be extended to higher dimensions, such as two-dimensional (2D) random walks, where the walker can move in any direction on a plane.

```python
import random
#using traditional ways
# def get_random_numbers(number, seed):
#     random.seed(seed)
#     random_numbers = []

#     for _ in range(number):
#         random_numbers.append(random.random())

#     return random_numbers

def get_random_numbers(seed, num_iterations):
    random.seed(seed)
    return [random.random() for _ in range(num_iterations)]  #using list comprehension
```

```python
step_size = 1
displacement = 0
displacement_list = []  # Initialize an empty list to store displacements
random_numbers = get_random_numbers(82697534,1000)

for num in random_numbers:
    if num > 0.5:
        displacement += step_size
    else:
        displacement -= step_size

    displacement_list.append(displacement)

# Print or use the displacement_list as needed
print(displacement_list)
```

```python
import matplotlib.pyplot as plt
# Plot the displacement values
plt.plot(displacement_list)
plt.xlabel('Step')
plt.ylabel('Displacement')
plt.title('Random Walk Displacement')
plt.show()
```

```python
def simulate_random_walk(steps, step_size, seed):
    random_numbers = get_random_numbers(seed, steps)
    displacement = 0
    cumulative_displacement = []

    for num in random_numbers:
        if num > 0.5:
            displacement += step_size
        else:
            displacement -= step_size

        cumulative_displacement.append(displacement)

    return cumulative_displacement

# Parameters
steps = 100
step_size = 1
seed = 42

# Simulate random walk and calculate cumulative displacement
cumulative_displacement = simulate_random_walk(steps, step_size, seed)

# Plot the cumulative displacement
plt.plot(cumulative_displacement)
plt.xlabel('Step')
plt.ylabel('Cumulative Displacement')
plt.title('Random Walk Cumulative Displacement')
plt.show()
```
