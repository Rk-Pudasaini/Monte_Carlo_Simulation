# What is Random Walk?
A random walk is a mathematical concept that describes a path or trajectory formed by a sequence of random steps. Each step is taken in a random direction, often with equal probability of moving in different directions. Random walks are widely used to model various phenomena in fields such as physics, finance, biology, and computer science.

In its simplest form, a random walk can occur in one dimension (1D), where the walker moves along a straight line either to the left or right with equal probability at each step. This can be extended to higher dimensions, such as two-dimensional (2D) random walks, where the walker can move in any direction on a plane.

```python
import random
#using traditional ways
# def get_random_numbers(number, seed):
#     random.seed(seed)
#     random_numbers = []

#     for _ in range(number):
#         random_numbers.append(random.random())

#     return random_numbers

def get_random_numbers(seed, num_iterations):
    random.seed(seed)
    return [random.random() for _ in range(num_iterations)]  #using list comprehension
```

```python
import matplotlib.pyplot as plt
# Plot the displacement values
plt.plot(displacement_list)
plt.xlabel('Step')
plt.ylabel('Displacement')
plt.title('Random Walk Displacement')
plt.show()
```

```python
def simulate_random_walk(steps, step_size, seed):
    random_numbers = get_random_numbers(seed, steps)
    displacement = 0
    cumulative_displacement = []

    for num in random_numbers:
        if num > 0.5:
            displacement += step_size
        else:
            displacement -= step_size

        cumulative_displacement.append(displacement)

    return cumulative_displacement

# Parameters
steps = 100
step_size = 1
seed = 42

# Simulate random walk and calculate cumulative displacement
cumulative_displacement = simulate_random_walk(steps, step_size, seed)

# Plot the cumulative displacement
plt.plot(cumulative_displacement)
plt.xlabel('Step')
plt.ylabel('Cumulative Displacement')
plt.title('Random Walk Cumulative Displacement')
plt.show()
```

#Algorithm
In the Monte Carlo simulation of a random walk, the following steps are typically involved:
1. Initialization: Start at a given initial state or position.
2. Random Step Generation: At each time step, randomly select a direction or state
transition according to a predefined probability distribution. This could involve moving left
or right, jumping to neighboring states in a lattice, or any other possible transition.
3. Update Position: Move the object to the new state based on the randomly generated
step.
4. Repeat: Iterate the process for a predetermined number of steps or until a certain
condition is met.
5. Data Collection: Record the position of the object at each time step to analyze the
behavior of the random walk.

```python
import numpy as np
def random_walk(steps):
    position = 0
    positions = [position]
    for _ in range(steps):
        step = np.random.choice([-1, 1])  # Randomly choose direction: -1 (left) or 1 (right)
        position += step
        positions.append(position)
    return positions

# Number of steps in the random walk
steps = 1000

# Perform random walk
positions = random_walk(steps)

# Plot the random walk
plt.plot(positions)
plt.title('1D Random Walk')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.show()
```

#Demonstration of 2D Random Walk

```python
def random_walk_2d(steps):
    # Initial position at the origin (0, 0)
    x_position, y_position = 0, 0
    x_positions, y_positions = [x_position], [y_position]

    # Generate random steps in x and y directions separately
    for _ in range(steps):
        x_step = np.random.choice([-1, 1])  # Randomly choose direction in x-axis: -1 (left) or 1 (right)
        y_step = np.random.choice([-1, 1])  # Randomly choose direction in y-axis: -1 (down) or 1 (up)

        # Update positions
        x_position += x_step
        y_position += y_step

        # Append positions to lists
        x_positions.append(x_position)
        y_positions.append(y_position)

    return x_positions, y_positions

# Number of steps in the random walk
steps = 1000

# Perform random walk in 2D
x_positions, y_positions = random_walk_2d(steps)

# Plot the random walk with marker points
plt.plot(x_positions, y_positions, marker='o', markersize=1, linestyle='-')
plt.title('2D Random Walk ')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.grid(True)
plt.show()
```

```python
import numpy as np

def transition_probs(num_positions):
    """
    Define transition probabilities for a 1D random walk.
    """
    probs = np.zeros((num_positions, num_positions))
    for i in range(num_positions):
        if i == 0:
            probs[i, i+1] = 1  # Move right
        elif i == num_positions - 1:
            probs[i, i-1] = 1  # Move left
        else:
            probs[i, i-1] = 0.5  # Move left
            probs[i, i+1] = 0.5  # Move right
    return probs

def random_walk_markov_chain(num_steps, num_positions):
    """
    Perform a random walk using Markov chain approach.
    """
    # Initialize position
    position = np.random.randint(num_positions)
    positions = [position]

    # Transition probabilities
    probs = transition_probs(num_positions)

    # Perform random walk
    for _ in range(num_steps):
        position = np.random.choice(range(num_positions), p=probs[position])
        positions.append(position)

    return positions

# Number of steps and positions
num_steps = 1000
num_positions = 21  # Number of positions on the line

# Perform random walk using Markov chain
positions = random_walk_markov_chain(num_steps, num_positions)

# Plot the random walk
import matplotlib.pyplot as plt
plt.plot(positions, marker='.', linestyle='-')
plt.title('1D Random Walk using Markov Chain')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def transition_probs_2d(num_positions):
    """
    Define transition probabilities for a 2D random walk.
    """
    probs = np.zeros((num_positions, num_positions))
    side_length = int(np.sqrt(num_positions))
    for i in range(num_positions):
        x, y = i % side_length, i // side_length
        if x > 0:
            probs[i, i-1] = 0.25  # Move left
        if x < side_length - 1:
            probs[i, i+1] = 0.25  # Move right
        if y > 0:
            probs[i, i-side_length] = 0.25  # Move down
        if y < side_length - 1:
            probs[i, i+side_length] = 0.25  # Move up

    # Normalize probabilities for each position
    probs /= np.sum(probs, axis=1, keepdims=True)

    return probs

def random_walk_markov_chain_2d(num_steps, num_positions):
    """
    Perform a 2D random walk using Markov chain approach.
    """
    # Initialize position
    position = np.random.randint(num_positions)
    positions = [position]

    # Transition probabilities
    probs = transition_probs_2d(num_positions)

    # Perform random walk
    for _ in range(num_steps):
        position = np.random.choice(range(num_positions), p=probs[position])
        positions.append(position)

    return positions

# Number of steps and positions
num_steps = 1000
side_length = 20  # Side length of the grid
num_positions = side_length * side_length  # Total number of positions on the grid

# Perform random walk using Markov chain
positions = random_walk_markov_chain_2d(num_steps, num_positions)

# Plot the random walk
# Convert positions to (x, y) coordinates
x_positions = [pos % side_length for pos in positions]
y_positions = [pos // side_length for pos in positions]

plt.figure(figsize=(8, 6))
plt.plot(x_positions, y_positions, marker='.', linestyle='-')
plt.title('2D Random Walk using Markov Chain')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.grid(True)
plt.show()
```
